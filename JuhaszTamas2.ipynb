```python
import os
ldir = os.chdir(r'C:\Users\Dell 5300\PorabaAnakonda')
import pandas as pd
import numpy as np

#1.feladat
df = pd.read_excel('Ertekesites(ETL).xlsx','hiányos',parse_dates=['Dátum'])
#2.feladat 
df[['Egységár','Érték']]=df[['Egységár','Érték']].astype(np.float).astype('Int32')
#4.feladat
df.head()
df.Terméknév= df.Terméknév.replace(['Tol','Tolll'],'Toll')
df.Terméknév = df.Terméknév.replace(['creuza','cerza'],'Ceruza')
#5.feladat
#a
a=[]
for row in df.itertuples():
    if  row.Mennyiség < 0:
        a.append(abs(row.Mennyiség))
    else:
        a.append(row.Mennyiség)
df.Mennyiség=a
#b
b=[]
for row in df.itertuples():
    if  row.Mennyiség > 200:
        b.append(None)
    else:
        b.append(row.Mennyiség)
df.Mennyiség=b
df['Mennyiség']=df['Mennyiség'].interpolate()
#3.feladat
#a
df = df.dropna(subset=['Mennyiség'])
#b
df = df.fillna({'Kategórianév':'ismeretlen'})
#c
dgg = df.groupby('Terméknév')['Terméknév','Egységár'].agg('mean')
y=[]
df['index'] = range(0,len(df))
df=df.fillna({'Egységár':0})
print(dgg)
print(df.head(15))
for row in df.itertuples():
    if  row.Egységár == 0:
        y.append(dgg.loc[row.Terméknév,'Egységár'])
    else:
        y.append(row.Egységár)
df.Egységár=y

print(df.head(15))
#d
x=[]
for row in df.itertuples():
    x.append(row.Egységár*row.Mennyiség)
df=df.drop(columns=['Érték'])
df['Bevétel']=x
#6.feladat
#df = df[df.duplicated(subset=['Kategórianév','Dátum','Terméknév','Bevétel','Vevőnév'],keep=False)]
df = df.sort_values('index').drop_duplicates(['Kategórianév','Dátum','Terméknév','Bevétel','Vevőnév'], keep='last')
#7.feladat
df.to_csv(r'Javított.csv', encoding="ISO-8859-2",sep = ',',index = None)

```
