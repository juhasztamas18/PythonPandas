```python
import os
ldir = os.chdir(r'C:\Users\Dell 5300\PorabaAnakonda')
import pandas as pd
import numpy as np
import time 
from datetime import datetime as dt
%matplotlib inline
import matplotlib.pyplot as plt


#1.feladat
df = pd.read_excel('Ertekesites(nagyker).xlsx','Lekérdezés1')

#2.feladat
df.groupby('Bolt város')['Nyereség'].mean().plot(kind='barh',color="green")

#3.feladat
df.groupby(['Vevőnév','Kategórianév'])['Nyereség'].count().unstack().plot(kind='bar',stacked=True)

#4.feladt/itt nem teljesen értettem a feladatott
ma=datetime.strptime(time.strftime("%Y.%m.%d"),"%Y.%m.%d")
## ha azt akarjuk, hogy az utolso vásárlás óta hány nap telt el
legutolso = ma - df['Dátum'].max()
##itt meg azt számoltam ki,hogy az aktuális vásárlástól
df['Eltelt_napok'] = ma - df['Dátum']

```

```python
#5.feladat itt külön kellett tennem mert össze folyt a két diagram
##df.groupby('Dátum')['Bevétel'].sum().plot(kind='pie',autopct='%.2f%%')
df.set_index('Dátum')
df['Nap'] =  df['Dátum'].dt.day_name()
df.groupby('Nap')['Bevétel'].sum().plot(kind='pie',autopct='%.2f%%')
```

```python
#6.feladat
d1 = np.random.randint(100,200,size=30)
d2 = np.random.randint(120,180,size=30)
df2 = pd.DataFrame()
szept = pd.date_range(start="2019-09-01",end='2019-09-30')
df2['Elso'] = d1
df2['Masodik'] = d2
df2['Dátum'] = szept
df2.set_index('Dátum')

plt.plot(df2['Dátum'], df2['Elso'],'r') 
plt.plot(df2['Dátum'], df2['Masodik'],'g') 
plt.xticks(fontsize=10, rotation=60)
plt.xlabel('Idö')
plt.ylabel('Értékek')
plt.title('Ábrázolás')
plt.legend(['Érték1', 'Érték2'])
plt.grid(True) # rácsozat
plt.show() # megjelenítés

```
