```python
import gzip
import pandas as pd
import json
from collections import defaultdict
import os
import numpy as np


# Folder Path
path = r'D:\Vortess\DWH\10_Datasets\LinkedIn Data'
  
# Change the directory
os.chdir(path)
r=0
x = set([])
#ldir = os.chdir(r'D:\Vortess\DWH\10_Datasets\LinkedIn Data')
alldf0 = pd.read_json(r'D:\Vortess\DWH\10_Datasets\LinkedIn Data\abdulla-9.json')
names = [x for x in alldf0 .columns]
alldf = pd.DataFrame(columns=names)
alldf["srcfrom"] = np.nan

# iterate through all file
for file in os.listdir():
    # Check whether file is in text format or not
    if file.endswith(".json"):
        try:
            r = r+1
            file_path = f"{path}\{file}"
            # call read text file function
            df = pd.read_json(file_path)
            df ["srcfrom"] = f"{file}"
            alldf = df.append(alldf)
            print(file_path + '| ' + str(r) + ' ' + file)
        except ValueError:
            pass  # do nothing!



```
