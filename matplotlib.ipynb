```python
x = np.arange(0,3*np.pi,0.1)
y2 = np.cos(x)
y3 = np.sin(x)
plt.plot(x, y2,'b') # koszinusz brzolsa kkkel
plt.plot(x, y3,'g') # szinusz brzolsa zlddel
plt.axhline(y=0, color = 'black') # viszintes tengely
plt.xlabel('x-tengely') # tengely felratok
plt.ylabel('y-tengely')
plt.title('cos s sin fggvnyek') # cm
plt.legend(['cos', 'sin']) # jelmagyarzat
plt.grid(True) # rcsozat
plt.show() # megjelents
```

```python
plt.bar([1,2,3],[5,2,3],width=0.3,color="red")
plt.xlabel("x tengely")
plt.ylabel("y tengely")
plt.title("Egyszer oszlopdiagram")
plt.show()
```

```python
x = np.random.rand(20)
y = np.random.rand(20)

plt.figure(figsize=(15, 15))

plt.subplot(321)
plt.scatter(x, y, s=80,  marker=">")

plt.subplot(322)
plt.scatter(x, y, s=80,  marker=(5, 0))

plt.subplot(323)
plt.scatter(x, y, s=80,  marker="o")

plt.subplot(324)
plt.scatter(x, y, s=80,  marker=(5, 1))

plt.subplot(325)
plt.scatter(x, y, s=80,  marker='+')

plt.subplot(326)
plt.scatter(x, y, s=80,  marker=(5, 2))

plt.show()
```

```python
x = np.arange(-5,6,1)
y1 = x
y2 = x**2
y3 = x**3
y4 = x**4

plt.figure(figsize=(15, 15))

plt.subplot(221)
plt.plot(x, y1)
plt.grid(True)

plt.subplot(222)
plt.plot(x, y2)
plt.grid(True)

plt.subplot(223)
plt.plot(x, y3)
plt.grid(True)

plt.subplot(224)
plt.plot(x, y4)
plt.grid(True)

plt.show()
```
